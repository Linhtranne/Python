```python

a = int(input("Nhập số nguyên a: "))
b = int(input("Nhập số nguyên b: "))

print(f"Các số nguyên tố trong khoảng [{a}, {b}]:")

prime = []

for num in range(a, b + 1):
    if num < 2:
        continue
    
    la_nguyen_to = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            la_nguyen_to = False
            break
    
    if la_nguyen_to:
        prime.append(num)

if prime:
    print(", ".join(map(str, prime)))
else:
    print("Không có số nguyên tố nào.")
```
