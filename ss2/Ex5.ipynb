```python
n = int(input("Nhập vào một số nguyên dương "))
if n <= 1:
    print("Số không hợp lệ, vui lòng nhập số nguyên dương")
else:
    n1 = n
    n2 = 0
    while n > 0:
        digit = n % 10
        n2 = n2 * 10 + digit
        n //= 10
    if n1 == n2:
        print(f"{n1} là số đối xứng")
    else:
        print(f"{n1} không phải là số đối xứng")
```
