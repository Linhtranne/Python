```python
import pandas as pd
from pathlib import Path
import json

logs_dir = Path("logs")
all_records = []

for path in logs_dir.glob("log_*.json"):
    date_str = path.stem.split("_")[1]
    date_parsed = pd.to_datetime(date_str, format="%Y-%m%d")
    with path.open(encoding="utf-8") as f:
        data = json.load(f)
    df_log = pd.DataFrame(data)
    df_log["date"] = date_parsed
    all_records.append(df_log)

if all_records:
    df_logs_all = pd.concat(all_records, ignore_index=True)
    df_logs_all["duration"] = pd.to_numeric(df_logs_all["duration"], errors="coerce")
    report = df_logs_all.groupby("date")["duration"].sum().reset_index()
    report.to_csv("duration_report_by_date.csv", index=False)
```
