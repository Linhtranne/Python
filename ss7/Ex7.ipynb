```python
import pandas as pd

df7 = pd.DataFrame({
    "name": ["An", "Bình", "Chi", "Dũng"],
    "date": ["2025-01-01", "2025-01-02", "2025-02-15", "2025-03-20"]
})

df7["date"] = pd.to_datetime(df7["date"])
df7["day"] = df7["date"].dt.day
df7["month"] = df7["date"].dt.month
df7["year"] = df7["date"].dt.year
df7["name"] = df7["name"].astype("category")
```
