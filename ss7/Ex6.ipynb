```python
import pandas as pd

df1_6 = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "name": ["An", "Bình", "Chi", "Dũng"]
})

df2_6 = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "score": [7.5, 8.0, 9.0, 6.5]
})

merged = pd.merge(df1_6, df2_6, on="id")

df_age_6 = pd.DataFrame({
    "age": [20, 22, 21, 23]
})
concat_result = pd.concat([merged, df_age_6], axis=1)
```
