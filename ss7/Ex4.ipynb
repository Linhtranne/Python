```python
import pandas as pd

df4 = pd.DataFrame({
    "name": ["An", "Bình", "Chi", "Dũng", "Bình"],
    "age": [19, 22, 20, 25, 24],
    "score": [8.5, 7.0, 6.5, 8.5, 7.0]
})

df4["age_group"] = df4["age"].apply(lambda x: "Young" if x <= 20 else "Adult")

score_map = {8.5: "Excellent", 7.0: "Good", 6.5: "Average"}
df4["score"] = df4["score"].map(score_map)

df4["score_int"] = df4["score"].map({"Excellent": 9, "Good": 8, "Average": 7}).astype(int)

df4["name"] = df4["name"].replace("Bình", "Bình (Updated)")
```
