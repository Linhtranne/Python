```python
import numpy as np
arr1 = np.array ([[88, 35, 62], [11, 98, 34], [64, 24, 48]])
print (f"Min" + str(np.min(arr1)))
print (f"Max" + str(np.max(arr1)))
print (f"Sum" + str(np.sum(arr1)))
print (f"Mean" + str(np.mean(arr1)))
print (f"Std" + str(np.std(arr1)))
print(f"Sum of colums: " + str(np.sum(arr1, axis=0)))
```
