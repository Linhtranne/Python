```python
import numpy as np
np.random.seed(42)
A = np.array([[5, 6, 8],
              [7, 2, 4],
              [1, 6, 3]])

print(A)
B = np.array([10, 20, 30])
print(B)
result_add = A + B
print("Kết quả phép cộng A + B:")
print(result_add)

result_sub = A - B
print("Kết quả phép trừ A - B:")
print(result_sub)
```
