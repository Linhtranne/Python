```python
import numpy as np


maxtrix = np.array([[6,9,7],[2,1,5],[7,8,3]])

if np.linalg.det(maxtrix) != 0:
    maxtrix_inv = np.linalg.inv(maxtrix)
    print("\nInverse matrix:\n", maxtrix_inv)
else:
    print("Matrix is not invertible.")

eig_values, eig_vectors = np.linalg.eig(maxtrix)

print("\nEigenvalues:", eig_values)
print("\nEigenvectors:\n", eig_vectors)
```
