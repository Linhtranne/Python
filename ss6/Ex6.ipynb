```python
import numpy as np

arr = np.array([[3, 7, 4, 5],[3, 5, 2, 4],[6, 3, 9, 7],[4, 8, 3, 5]])

print(f"sum of column: {arr.sum(axis=0)}")
print(f"moan of column: {arr.mean(axis=0)}")
print(f"std of column: {arr.std(axis=0)}")

print(f"sum of row: {arr.sum(axis=1)}")
print(f"moan of row: {arr.mean(axis=1)}")
print(f"std of row: {arr.std(axis=1)}")
```
