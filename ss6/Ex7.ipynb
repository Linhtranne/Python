```python
import numpy as np
from datetime import datetime

start = datetime.now()
arr1 = np.array([[ 8, 12, 14,  6,  4],[18, 13,  2,  2, 12],[ 6, 18,  5, 10, 10],[ 9,  2,  4, 12, 12],[ 2,  3,  2, 17,  3]])
arr2 = np.array([10, 20, 30, 40, 50])

print(f"total: {arr1 + arr2}")
print(f"diff: {arr1 - arr2}")
print(f"root: {arr1**(1/2)}")
print(f"sum 5: {arr1 + 5}")

end = datetime.now()
print(f"Time to done: {end - start}")
```
