```python
import numpy as np
import time
np.random.seed(42)
arr = np.random.randint(1, 21, size=10)
print("Mảng ban đầu:")
print(arr)
squared = arr ** 2
print(squared)
sqrt_arr = np.sqrt(arr)
print(sqrt_arr)
result_add = np.add(arr, 10)
print(result_add)
result_multiply = np.multiply(arr, 2)
print(result_multiply)
result_sqrt = np.sqrt(arr)
print(result_sqrt)
print("=" * 50)
print("SO SÁNH TỐC ĐỘ: Vectơ hóa vs Vòng lặp")
print("=" * 50)
large_array = np.random.randint(1, 100, size=(1000, 1000))
start_time = time.time()
result_vectorized = large_array ** 2
time_vectorized = time.time() - start_time
start_time = time.time()
result_loop = np.zeros_like(large_array)
for i in range(large_array.shape[0]):
    for j in range(large_array.shape[1]):
        result_loop[i, j] = large_array[i, j] ** 2
time_loop = time.time() - start_time
print(f"Thời gian vectơ hóa (NumPy): {time_vectorized:.6f} giây")
print(f"Thời gian vòng lặp thủ công: {time_loop:.6f} giây")
print(f"Nhanh hơn: {time_loop / time_vectorized:.2f} lần")
```
