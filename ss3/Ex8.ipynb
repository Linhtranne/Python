```python
def isPerfectNumber(n):
    if n <= 1:
        return False
    
    sum_divisors = 0
    for i in range(1, n):
        if n % i == 0:
            sum_divisors += i
    
    return sum_divisors == n

n = int(input("Nhập một số nguyên: "))

if isPerfectNumber(n):
    print(f"{n} là số hoàn hảo")
else:
    print(f"{n} không phải là số hoàn hảo")
```
