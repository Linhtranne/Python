```python
import numpy as np
import matplotlib.pyplot as plt
months = list(range(1, 13))
temperature = [20, 22, 25, 28, 30, 32, 31, 30, 28, 26, 23, 21]
rainfall = [10, 15, 30, 120, 200, 250, 280, 220, 150, 80, 30, 15]
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(months, temperature, color='red', linestyle='-', linewidth=2, 
        marker='o', markersize=6, label='Nhiệt độ trung bình (°C)')
ax.scatter(months, rainfall, color='blue', s=80, alpha=0.7, 
           label='Lượng mưa (mm)', zorder=5)
ax.set_title("Nhiệt độ và lượng mưa Hà Nội 2025", fontsize=14, fontweight='bold')
ax.set_xlabel("Tháng", fontsize=12)
ax.set_ylabel("Giá trị", fontsize=12)
ax.set_xticks(months)
ax.grid(True, alpha=0.3)
ax.legend(loc='upper left', fontsize=10)
plt.show()
```
