```python
import csv

with open('students.csv', mode='r', encoding='utf-8') as infile:
    reader = csv.reader(infile)
    header = next(reader)

    high_grade_students = []

    for row in reader:
        name, age, grade = row[0], row[1], float(row[2])
        if grade >= 7:
            high_grade_students.append(row)
            print(f"Sinh viên: {name}, Tuổi: {age}, Điểm: {grade}")

with open('high_grade_students.csv', mode='w', newline='',
          encoding='utf-8') as outfile:
    writer = csv.writer(outfile)
    writer.writerow(header)
    writer.writerows(high_grade_students)
```
