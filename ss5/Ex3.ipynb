```python
import json

with open('user_data.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

person = next((item for item in data if item.get('id') == 1), None)
if person:
    print(f"Tên: {person.get('name')}, Tuổi: {person.get('age')}")
else:
    print("Không tìm thấy người có id = 1")

new_user = {
    "id": 2,
    "name": "linhtranne",
    "age": 20
}

data.append(new_user)

with open('user_data.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=4)
```
