```python

my_list = [1, 2, 3, 2, 5, 1, 7, 9, 9, 4]
print("List ban đầu:", my_list)
print()

even_positions = my_list[::2]
print("Các vị trí chẵn:", even_positions)

my_tuple = tuple(my_list)
print("Tuple:", my_tuple)

my_dict = {i: value for i, value in enumerate(my_list)}
print("Dict:", my_dict)

my_set = set(my_list)
print("Set:", my_set)

duplicates_removed = len(my_list) - len(my_set)
print("Số phần tử bị loại trùng:", duplicates_removed)
```
