```python
import numpy as np
lists = np.random.randint(0, 10, size=10)
print("Danh sách ban đầu:", lists)
print("Các vị trí chẵn:", lists[::2])
print("Tuple" + str(tuple(lists[::2])))
print("Dict"+ str({i: lists[i] for i in range(0, len(lists), 2)}))
print("Set" + str(set(lists[::2])))
print("Số phần tử bị loại trùng:", len(lists[::2]) - len(set(lists[::2])))
```
