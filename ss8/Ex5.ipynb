```python
import pandas as pd

students_df = pd.DataFrame({
    "id": [1, 2, 3, 4, 5],
    "name": ["An", "Binh", "Chi", "Duy", "Hanh"]
})
students_df.to_csv("students.csv", index=False)

scores_df = pd.DataFrame({
    "id": [1, 1, 2, 2, 3, 3, 3, 4, 5, 5],
    "score": [7.5, 8.0, 6.5, 7.0, 9.2, 8.8, 7.9, 6.8, 8.5, 9.0]
})
scores_df.to_csv("scores.csv", index=False)

students = pd.read_csv("students.csv")
scores = pd.read_csv("scores.csv")

merged = pd.merge(students, scores, on="id")

report = (
    merged
    .groupby(["id", "name"])["score"]
    .agg(min_score="min", max_score="max", mean_score="mean")
    .reset_index()
)


report.to_csv("report.csv", index=False)

print("Merged data:")
print(merged)
print("\nReport (min / max / mean):")
print(report)
```
