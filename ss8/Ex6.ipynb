```python
import numpy as np


a = np.array([[1], [2], [3], [4]]) 


b = np.array([[10, 20, 30, 40, 50]]) 

print("a:\n", a)
print("shape a:", a.shape)
print("\nb:\n", b)
print("shape b:", b.shape)

c = a * b
print("\nKết quả c = a * b:\n", c)
print("shape c:", c.shape)


c_sqrt = np.sqrt(c)
c_log = np.log(c) 

print("\n√c:\n", c_sqrt)
print("\nlog(c):\n", c_log)

c_min = np.min(c, axis=1)
c_max = np.max(c, axis=1)
c_mean = np.mean(c, axis=1)
c_std = np.std(c, axis=1)

print("\nGiá trị theo từng hàng (axis=1):")
print("min :", c_min)
print("max :", c_max)
print("mean:", c_mean)
print("std :", c_std)
```
