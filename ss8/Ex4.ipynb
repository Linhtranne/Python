```python
import pandas as pd
name = ['An', 'Bình', 'Châu', 'Dũng', 'Hà']
age = [20, 21, 22, 20, 23]
score = [7.5, 8.0, 6.5, 9.0, 8.5]
score_series = pd.Series(score, index=name)
print("Series điểm sinh viên:")
print(score_series)
df = pd.DataFrame({
    'name': name,
    'age': age,
    'score': score
})
print(df)
df.info()
print(df.describe())
print("Sinh viên >= 8 điểm:")
filtered_df = df.loc[df['score'] >= 8]
print(filtered_df.values.tolist())
```
