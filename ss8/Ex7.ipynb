```python
import numpy as np
import pandas as pd

data = {
    "name": ["An", "Bình", "Châu", "Dũng"],
    "age": [20, np.nan, 22, 21],
    "score": [8.5, np.nan, 7.0, 9.0],
    "city": ["Hà Nội", "HCM", np.nan, "Hà Nội"]
}

df = pd.DataFrame(data)
print("Data gốc:")
print(df)

print("\nKiểm tra dữ liệu thiếu (df.isna()):")
print(df.isna())

age_mean = df["age"].mean()
score_median = df["score"].median()

df["age"] = df["age"].fillna(age_mean)
df["score"] = df["score"].fillna(score_median)
df["city"] = df["city"].fillna("Unknown")

print("\nData sau khi fillna:")
print(df)

df["city"] = df["city"].astype("category")
print("\nKiểu dữ liệu các cột:")
print(df.dtypes)

def score_level(x: float) -> str:
    if x >= 8:
        return "Good"
    elif x >= 7:
        return "Average"
    else:
        return "Bad"

df["score_level"] = df["score"].apply(score_level)
print("\nData sau khi thêm cột score_level:")
print(df)

age_mean = df["age"].mean()
age_std = df["age"].std()

df["age_normalized"] = (df["age"] - age_mean) / age_std
print("\nData sau khi chuẩn hóa tuổi (age_normalized):")
print(df)
```
